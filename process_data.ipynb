```python
import numpy as np
import pandas as pd
```

```python
stations = pd.read_csv('./data/underground-stations.csv').fillna('')
routes = pd.read_csv('./data/underground-routes.csv').fillna('')
travel_times = pd.read_csv('./data/underground-travel-time.csv').fillna('')
```

```python
stations = stations.set_index('id')
stations.index.names = [None]
routes = routes.set_index('line')
routes.index.names = [None]
```

```python
stations
```

```python
data = {}
for k, v in stations.iterrows():
    data[k] = {'id': k,
               'latitude': v['latitude'],
               'longitude': v['longitude'],
               'name': v['name'],
               'zone': v['zone'],
               'total_lines': v['total_lines'],
               'rail': v['rail']}
```

```python
data
```

```python
for k in data.keys():
    data[k]['destinations'] = []
```

```python
for row in travel_times.iterrows():
    data[row[1]['station1']]['destinations'].append({'station_id': row[1]['station2'], 
                                                     'name': stations['name'][row[1]['station2']],
                                                     'line_id': row[1]['line'], 
                                                     'line_name': routes['name'][row[1]['line']],
                                                     'line_colour': routes['colour'][row[1]['line']],
                                                     'line_stripe': routes['stripe'][row[1]['line']],
                                                     'time': row[1]['time'],
                                                     'latitude': data[row[1]['station2']]['latitude'],
                                                     'longitude': data[row[1]['station2']]['longitude']})
    data[row[1]['station2']]['destinations'].append({'station_id': row[1]['station1'], 
                                                     'name': stations['name'][row[1]['station1']],
                                                     'line_id': row[1]['line'], 
                                                     'line_name': routes['name'][row[1]['line']],
                                                     'line_colour': routes['colour'][row[1]['line']],
                                                     'line_stripe': routes['stripe'][row[1]['line']],
                                                     'time': row[1]['time'],
                                                     'latitude': data[row[1]['station1']]['latitude'],
                                                     'longitude': data[row[1]['station1']]['longitude']})                                               
```

```python
data_flat = []
for k, v in data.items():
    s = {'id': k,
         'zone': v['zone'],
         'total_lines': v['total_lines'],
         'destinations': v['destinations'],
         'name': v['name'],
         'rail': v['rail'],
         'longitude': v['longitude'],
         'latitude': v['latitude']}
    data_flat.append(s)
```

```python
import json
json.dump(data_flat, open('data.json', 'w'))
```
